### 1. Write a function that calculates the dot product manually. It should accept two vectors and perform that calculation. Be sure to use try except and error handling for basic issues that might occur.

```python
x=[3,3,3]
y=[4,4,5]
def dotproduct(x,y):
    summed=0
    if len(x)==len(y) and len(x)!=0:
        try:
            for i in range(len(x)):
                summed=summed + sum([x[i]*y[i]])
        except IndexError:
                print("Arrays should be the same size.")
        except TypeError:
                print("All input values should be numbers.")
        except Exception as e:
                return e
        else:
                return summed

print(dotproduct(x,y))
```

### 2. In a markdown section in your notebook, explain what one hot encoding is and why we use it. Then, manually calculate a one hot encoding for the Belize_Airbnb_listings.csv neighborhood_group column. Note: manually means donâ€™t use scikitlearn or keras.

Onehot encoding is a coded way to convert categorical data to numerical data (binary vectors). Onehot encoding helps us analyze string data for patterns more easily and quickly (because processes run faster with numerical/binary data).

```python
belize_nbg_list = belize_df.drop_duplicates(subset=["neighbourhood_group"])
#print(belize_nbg_list["neighbourhood_group"]): Belize, Stann Creek, Toledo, Cayo, Corozal, and Orange Walk.

#convert neighbourhood_group unique values to numbers

### Categorical data to be converted to numeric data
nb_groups_df = belize_df["neighbourhood_group"]

### List of neighbourhood groups
nb_groups = ["Belize", "Stann Creek", "Toledo", "Cayo", "Corozal", "Orange Walk"]

### map each neighbourhood group to an integer
mapping = {}
for x in range(len(nb_groups)):
  mapping[nb_groups[x]] = x

nbg_one_hot_encoded = []

for n in nb_groups_df:
  arr = list(np.zeros(len(nb_groups), dtype = int))
  arr[mapping[n]] = 1
  nbg_one_hot_encoded.append(arr)

print(nbg_one_hot_encoded)
```
